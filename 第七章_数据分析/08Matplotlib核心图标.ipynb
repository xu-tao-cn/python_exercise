```python
import random

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from wcwidth import width

# 展示中文
plt.rcParams['font.sans-serif'] = ['SimHei']  #如果有中文则展示为黑体
# Figure是画布 any是子图数组 nrows是行 ncols是列 figsize是大小 dpi是清晰度
Figure, any=plt.subplots(nrows=1, ncols=2,figsize=(10,5),dpi=100)

# 图一:柱状图
country = ['中国', '美国', '日本', '印度', '法国', '英国', '俄罗斯', '意大利', '德国', '西班牙']
people = [random.randint(10000000, 20000000) for i in range(10)]
any1: plt.Axes = any[0]
any1.set_title('柱状图')
any1.set_xlabel('国家')
any1.set_ylabel('人口数')
any1.bar(country, people,width=0.6,color='green')
any1.grid(linestyle = '-.',alpha=0.4)

# 图二:饼状图
people_kinds = ['中国', '美国', '日本', '印度', '法国', '英国', '俄罗斯', '意大利', '德国', '西班牙']
people_GGSize = [random.randint(10, 30) for i in range(10)]

any2: plt.Axes = any[1]
any2.set_title('饼状图')
# autopct表示显示百分比, shadow表示阴影
any2.pie(people_GGSize, labels=people_kinds, autopct='%1.1f%%', shadow=False)
any2.legend(loc='best',ncols=5, bbox_to_anchor=(1.3, -0.01))

plt.savefig('./Data/08Matplotlib核心图标.png')
plt.show()
```
