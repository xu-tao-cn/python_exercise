```python
import pandas as pd

# 读取数据
df = pd.read_csv("./Data/sales.csv",usecols=['订单号','产品名称','产品类别','单价','销售数量'],index_col='订单号')

df['销售额'] = df['单价'] * df['销售数量']

df.to_csv("./Data/sales_result.csv",index=False)

df.head(10)
df.tail(10)
df.describe()
df.info
# 一个中括号取一个列
df['产品名称']  
# 两个中括号取多个列
df[['产品名称','单价']]
```

```python
# 选择行iloc, loc
# 示例:df.iloc[start:end:step,start:end:step] 表示: 开始:结束:步长
# iloc基于行号进行选择
df.iloc[0:100:2]

# 基于索引进行选择
df.loc[6805677496:5852807025:2]
```

```python
# 获取产品类别是食品或服装的
df[df['产品类别'].isin(['食品','服装'])]
# 获取销售数量大于10的且单价大于400的
df[(df['销售数量'] >= 10) & (df['单价'] >= 400)]
```
