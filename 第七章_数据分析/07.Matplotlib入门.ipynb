```python
import random

import matplotlib.pyplot as plt


# 展示中文
plt.rcParams['font.sans-serif'] = ['SimHei']  #如果有中文则展示为黑体

# 绘制折线图
x = [i for i in range(1, 15)]
y_1 = [random.randint(1, 100) for i in x]
y_2 = [random.randint(1, 100) for i in x]

# 画布大小设置
plt.figure(figsize=(10, 5))
# 绘制折线图
plt.plot(x, y_1, label='百香果')
plt.plot(x, y_2, label='西瓜')
# 设置标题
plt.title('折线图')
# 添加x,y轴标签
plt.xlabel('时间')
plt.ylabel('销量')
# 添加图例
plt.legend(loc='best')
# 设置x轴刻度
plt.xticks(x[::2])
# 设置y轴刻度
plt.yticks(range(0, 101, 5))
# plt.xticks(x)
# 设置网格
plt.grid(linestyle = '--', alpha = 0.3)
plt.show()
```
